```python
import os
import dotenv
dotenv.load_dotenv()
API_HOST=os.environ['API_HOST']
API_TOKEN='{}'.format(os.environ['API_TOKEN'])
PREFIX = 'Bearer'
print('Using host {}'.format(API_HOST))
```

```python
import conquest as c
config = c.Configuration()
config.host = API_HOST
config.api_key['Authorization'] = API_TOKEN
config.api_key_prefix['Authorizations'] = PREFIX
api_instance = c.ApiClient(configuration=config)
system = c.SystemServiceApi(api_instance).application_version()
print(system)
```

```python
asset = c.AssetServiceApi(api_instance).get_asset(0)
print(asset.to_dict()['action_permission'])
```
